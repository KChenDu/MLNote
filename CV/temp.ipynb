```python
import math

def bivariate_normal(X,Y,sigma_X=1,sigma_Y=1,mu_X=0,mu_Y=0,sigma_XY=0):
    rho = sigma_XY/(sigma_X*sigma_Y)
    z = np.square(X-mu_X)/sigma_X**2\
        - 2*rho*(X-mu_X)*(Y-mu_Y)/(sigma_X*sigma_Y)\
        + np.square(Y-mu_Y)/sigma_Y**2
    pdf = 1/(2*math.pi*sigma_X*sigma_Y*math.sqrt(1-rho**2))\
        *np.exp(-z/(2*(1-rho**2)))
    return pdf
    
delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y) # define mesh of points
Z1 = bivariate_normal(X, Y, 1.0, 1.0, 0.0, 0.0)
Z2 = bivariate_normal(X, Y, 1.5, 0.5, 1, 1)
Z = 10.0 * (Z2 - Z1) # difference of Gaussians

# Create a simple contour plot with labels using default colors.  The
# inline argument to clabel will control whether the labels are draw
# over the line segments of the contour, removing the lines beneath
# the label
plt.figure()
CS = plt.contour(X, Y, Z)
plt.clabel(CS, inline=1, fontsize=10)
plt.title('Simplest default with labels')
plt.show()
```
