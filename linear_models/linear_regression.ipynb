# [线性回归](linear_regression.ipynb)
$$f(\mathbf x_i) = \mathbf \omega^\top\mathbf x_i + b\Rightarrow f(\mathbf X) =\mathbf X\hat{\mathbf \omega}$$
$$\hat{\mathbf \omega}\ast = \argmin_{\hat{\mathbf \omega}}(\mathbf y - \mathbf X\hat{\mathbf \omega})^\top(\mathbf y - \mathbf X\hat{\mathbf \omega})$$
$$\frac{\partial E_{\hat{\mathbf \omega}}}{\partial \hat{\mathbf \omega}} = 2\mathbf X^\top(\mathbf X\hat{\mathbf \omega} - \mathbf y)\Rightarrow \hat{\mathbf \omega}\ast = (\mathbf X^\top\mathbf X)^{-1}\mathbf X^\top\mathbf y \Rightarrow f(\hat{\mathbf x}_i) = \hat{\mathbf x}_i^\top(\mathbf X^\top\mathbf X)^{-1}\mathbf X^\top\mathbf y$$
- 对数线性回归：$\ln y = \mathbf\omega\mathbf x + b$
- 广义线性模型：$y = g^{-1}(\mathbf\omega\mathbf x + b)$

```python
import numpy as np
import matplotlib.pyplot as plt

from pathlib import Path
from sklearn.linear_model import LinearRegression


IMAGES_PATH = Path() / "images" / "training_linear_models"
IMAGES_PATH.mkdir(parents=True, exist_ok=True)


def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = IMAGES_PATH / f"{fig_id}.{fig_extension}"
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)


X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

plt.figure(figsize=(6, 4))
plt.plot(X, y, "b.")
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)
plt.axis([0, 2, 0, 15])
plt.grid()
save_fig("generated_data_plot")
plt.show()
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
lin_reg.intercept_, lin_reg.coef_
```

```python
X_new = np.array([[0], [2]])
y_predict = lin_reg.predict(X_new)
y_predict
```

```python
plt.figure(figsize=(6, 4))  # extra code – not needed, just formatting
plt.plot(X_new, y_predict, "r-", label="Predictions")
plt.plot(X, y, "b.")

# extra code – beautifies and saves Figure 4–2
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)
plt.axis([0, 2, 0, 15])
plt.grid()
plt.legend(loc="upper left")
save_fig("linear_model_predictions_plot")

plt.show()
```
